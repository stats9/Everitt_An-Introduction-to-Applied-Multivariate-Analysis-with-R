```r
# load packages
library(tidyverse)
library(corrplot)
library(sf)
library(igraph)

# define vertices
city_per <- c("شیراز", 
"اصفهان", "آمل", 
"بیرجند", "بجنورد", "مشهد", "سیرجان", "تهران", "یزد")


# define latitude and longitude for city
mydat <- data.frame(
    City = city_per, 
    Lat = c(29.591768, 32.661343, 36.471546, 
    32.83333, 37.4745, 36.310699, 29.45137, 35.7219, 31.8974), 
    Lon = c(52.583698, 51.680374, 52.355087, 59.221375, 
    57.32903, 59.599457, 55.6809, 51.3347, 54.3569)
)




# read shape file of Province map of Iran

iran <- st_read("D:\\habib_repository_github\\iran_data_map\\irn_admbnda_adm1_unhcr_20190514.shp")


# create map of iran

ggplot(iran) + geom_sf(fill = NA)  + 
theme_minimal() -> map_iran
map_iran +
geom_text(data = mydat, 
aes(x = Lon, y = Lat, label = City), color = "darkblue") -> P1
P1



```

```r
# create distance matrix 

dist_mat <- matrix(
    c(0, 489, 1110, 1073, 1344, 1354, 387, 920, 441, 489, 0, 609, 870, 
    1019, 1147, 627, 436, 323, 1110, 609, 0, 1045, 518, 802, 1147, 
    185, 799, 
    1073, 870, 1045, 0, 659, 492, 740, 1135, 628, 
    1344, 1019, 518, 659, 0, 
    274, 1183, 722, 906, 1354, 1147, 802, 492, 274, 0, 1085, 
    899, 907, 
    387, 627, 1147, 740, 1183, 1085, 0, 969, 351,
    920, 436, 185, 1135, 
    722, 899, 969, 0, 621, 441, 323, 799, 628, 906, 907, 351, 
    621, 0
    ), 9, 9, dimnames = list(mydat$City, mydat$City)
)
```

```r
check_symmetric_matrix <- function(A){
    n1 <- nrow(A)
    n2 <- ncol(A)
    if(n1 != n2) stop("A must be square matrix")
    S <- c()
    k <- 0
    for(i in 1:n1){
        for(j in 1:n2){
        s <- A[i, j] != A[j, i]
        S[k] <- s
        }
    }
    res <- ifelse(sum(S), "A not Symmetric", "A is Symmetric")
    return(res)
}
check_symmetric_matrix(dist_mat)
```

```r
# create a mat distance

corrplot(dist_mat, type = "full",
method = "number", 
is.corr = FALSE, col = c("green", "red", "darkblue"), 
number.digits = 0)
```

```r
# use igraph for visualize data and implement MST Algorithm 

G <- make_empty_graph(n = 9)

V(G)$color <- "skyblue2"
V(G)$size <- 20
coord <- cbind(lon = mydat$Lon, lat = mydat$Lat)
col1 <- c(rep(1, 8), rep(2, 7), 
rep(3, 6), rep(4, 5), rep(5, 4), rep(6, 3), 
rep(7, 2), 8)
col2 <- c(2:9, 3:9, 4:9, 5:9, 6:9, 7:9, 8:9, 9)
col3 <- c(dist_mat[2:9, 1], 
dist_mat[3:9, 2], dist_mat[4:9, 3], 
dist_mat[5:9, 4], dist_mat[6:9, 5], dist_mat[7:9, 6], 
dist_mat[8:9, 7], dist_mat[9, 8])
names(col1) <- names(col2) <- names(col3) <- NULL
Edge <- numeric(72)
Edge[seq(1, 72, 2)] <- col1
Edge[seq(2, 72, 2)] <- col2

G %>% 
  add_edges(Edge) %>%
  set_edge_attr("color", value = "red") -> G
E(G)$weight <- E(G)$label <- col3
V(G)$label = city_per
G <- set_graph_attr(G, "layout", coord)
G <- set_edge_attr(G, "color", value = 1:36)
G <- set_edge_attr(G, "label.color", value = 1:36)
plot.igraph(G, edge.arrow.mode = 0)
```

```r
G_mst <- mst(G, algorithm = "prim")
plot.igraph(G_mst, edge.arrow.mode = 0)
```
